```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import cv2
```

```python
excel_file_path = "gaze_data.xlsx"
image_file_path = "bg-image.png"
```

# New Section

```python
gaze_data = pd.read_excel("/content/gaze_data.xlsx")


background_img = cv2.imread("/content/bg-image.png")
background_img = cv2.cvtColor(background_img, cv2.COLOR_BGR2RGB)
```

```python
# Extract screen dimensions from the image
screen_height, screen_width, _ = background_img.shape

# Create a figure for the heatmap
fig, ax = plt.subplots(figsize=(12, 8))

# Plot the background image
ax.imshow(background_img, extent=[0, screen_width, screen_height, 0], origin="upper")
```

```python
# Plot the heatmap using gaze data
sns.kdeplot(
    x=gaze_data["Gaze_X"],
    y=gaze_data["Gaze_Y"],
    fill=True,
    cmap="RdBu_r",  # Colormap for the heatmap
    alpha=0.4,  # Transparency of the heatmap
    ax=ax,
)

# Customize the plot
ax.set_xticks([])
ax.set_yticks([])
ax.set_frame_on(False)
ax.set_xlim(0, screen_width)
ax.set_ylim(screen_height, 0)
```

```python
# Save and show the first heatmap
plt.savefig('heatmap_rdbr.png', bbox_inches='tight', pad_inches=0)
plt.show()

# Create the second heatmap (colormap: 'Dark2')
fig, ax = plt.subplots(figsize=(12, 8))

# Plot the background image
ax.imshow(background_img, extent=[0, screen_width, screen_height, 0], origin="upper")

# Plot the heatmap using gaze data
sns.kdeplot(
    x=gaze_data["Gaze_X"],
    y=gaze_data["Gaze_Y"],
    fill=True,
    cmap="Dark2",  # Colormap for the heatmap
    alpha=0.7,  # Transparency of the heatmap
    ax=ax,
)

# Customize the plot
ax.set_xticks([])
ax.set_yticks([])
ax.set_frame_on(False)
ax.set_xlim(0, screen_width)
ax.set_ylim(screen_height, 0)

# Save and show the second heatmap
plt.savefig('heatmap_dark2.png', bbox_inches='tight', pad_inches=0)
plt.show()
```
